# *************************** Lamda ************************* #(Anonymous Function)
# Syntax : Lamda arguments: expression

```python
def square(x):
    return x**2

print(square(4))
```

```python
sq_lambda = lambda x: x**2
```

```python
sq_lambda(6)
```

```python
add = lambda a,b: a+b
add(3,4)
```

```python
is_even = lambda x: x%2 == 0
is_even(21)
```

```python
words = ["Machine Learning", "Data Structures", "Python Basics"]
sorted_words = sorted(words, key = lambda x: len(x))

sorted_words
```

```python
fib = lambda n : n if n<=1 else fib(n-1) + fib(n-2)
[fib(i) for i in range(10)]
```

```python
#Factorial 
def fact_num(n):
    if n==0:
        return 1
    else:
        return n*fact_num(n-1)
```

```python
fact_num(5)
```

```python
fact_num = lambda n: 1 if n==0 else n*fact_num(n-1) 
fact_num(5)
```

# *********************** Map ********************* #(Executes a specified function for each item of an iterable )

```python
l = [1,2,3,4,5]

def square(l):
    sq = []
    for i in l:
        sq.append(i**2)
    return sq

square(l)
```

```python
def sq(x):    
    return x**2

sq(10)
```

```python
l
```

```python
map(sq,l)
```

```python
list(map(sq,l))  # Map function map each elements of list(l) to sq
```

```python
# Using Lamda with map
list(map(lambda x: x**2,l))  # Here we square value value of x and l is the list (we map list with x using map)
```

# ********************* Reduce ********************* #(Folding/Reduction)

```python
from functools import reduce
```

```python
l = [1,2,3,4,5,6,7,8]
reduce(lambda x,y: x+y,l)  # We prefer educe with 2 numbers only it will throw error with more or less number or empty list
```

```python
# Can be used for strings as well 
```
