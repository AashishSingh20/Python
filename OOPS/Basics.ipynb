```python
class Car:
    def accelerate(self):   # self represents the instance of the class(without it a.accelerate will give error)
        print("Car is Accelerating")  # We can write anything in place of self it will still work as self is not a python keyword
```

```python
class Bank:
    Balance = 3000
    def deposit(self,amount):
        print(f"I am depositing {amount} rupees")

    def withdraw(self,amount):
        print(f"I am Withdrawing {amount} rupees")
```

```python
class List:
    l = [1,2,3,4,5,6,7,8]
    def extract_even(self, l):
        l1 = []
        for i in l:
            if i%2 == 0:
                l1.append(i)
        
        return l1
```

```python
obj1 = List()
print(obj1.l)
obj1.extract_even(obj1.l)  
```

```python
class List2:

    def __init__(self,l):     # Dunder Method
        self.l = l
    def extract_even(self):
        l1 = []
        for i in self.l:
            if i%2 == 0:
                l1.append(i)
        
        return l1
```

```python
obj2 = List2([1,2,3,4,5,6,7,8])
```

```python
obj2.extract_even()
```

```python
class Book:
    def __init__(self, name, author, title):
        self.name_of_book = name
        self.book_author = author
        self.title_name = title

    def extract_details_name_title(self):
        print(self.name_of_book, self.title_name)

    def extract_details_name_author(self):
        print(self.name_of_book, self.book_author)

    
```

```python
student1 = Book("ML","John", "Clustering")

student1.extract_details_name_author()
student1.extract_details_name_title()
```

```python
class Book:
    def __init__(self):
        self.name_of_book = "DSA"
        self.book_author = "Luther"
        self.title_name = "DP"

    def extract_details_name_title(self):
        print(self.name_of_book, self.title_name)

    def extract_details_name_author(self):
        print(self.name_of_book, self.book_author)
```

```python
student2 = Book()
student2.extract_details_name_author()
student2.extract_details_name_title()
```
