# *********************** Generators ********************** #(A Type of type which does not return a single value, instead returns an iterator object)

```python
def square_numbers(n):
    result = []
    for i in range(n):
        result.append(i**2)
    return result

square_numbers(3)
    
```

```python
def square_numbers_generator(n):
    for i in range(n):
        yield i**2    # This generates one value at a time(Lazy but Memory Efficient)

gen = square_numbers_generator(4)
```

```python
def fibonacci_gen(n):
    a = 0
    b = 1
    for i in range(n):
        yield a
        a, b = b, a+b
```

```python
f = fibonacci_gen(1000000)
```

```python
next(f)
```

```python
next(f) # And so on Till you want to get answers
```
